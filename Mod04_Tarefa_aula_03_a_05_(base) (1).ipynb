## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
#0) Importar bibliotecas
import pandas as pd
import numpy as np

# Dados da quantidade de acertos 
np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.

def calcular_percentual_acerto(df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui):
    # Calcula o total de questões por disciplina
    total_questoes = {
        'Matemática': 24,
        'Português': 18,
        'Geografia': 8,
        'Inglês': 8,
        'História': 8,
        'Física': 12,
        'Química': 12
    }

    # Calcula o percentual de acerto por disciplina
    df_mat['percentual_acerto'] = (df_mat['Qt_acertos'] / total_questoes['Matemática']) * 100
    df_por['percentual_acerto'] = (df_por['Qt_acertos'] / total_questoes['Português']) * 100
    df_geo['percentual_acerto'] = (df_geo['Qt_acertos'] / total_questoes['Geografia']) * 100
    df_ing['percentual_acerto'] = (df_ing['Qt_acertos'] / total_questoes['Inglês']) * 100
    df_his['percentual_acerto'] = (df_his['Qt_acertos'] / total_questoes['História']) * 100
    df_fis['percentual_acerto'] = (df_fis['Qt_acertos'] / total_questoes['Física']) * 100
    df_qui['percentual_acerto'] = (df_qui['Qt_acertos'] / total_questoes['Química']) * 100

    # Retorna os DataFrames com os percentuais de acerto
    return df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui

# Exemplo de uso
df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui = calcular_percentual_acerto(df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui)

# Exemplo: Imprime o percentual de acerto em Matemática
print("Percentual de acerto em Matemática:")
print(df_mat['percentual_acerto'])
````

````python
# 1) Seu código aqui

df_materias = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)

# Crie um novo DataFrame concatenando os DataFrames das matérias
df_materias = pd.concat(materias, axis=1)

# Defina os nomes das colunas
df_materias.columns = ["Matemática", "% Acertos","Português", "% Acertos", "Geografia", "% Acertos", "Inglês", "% Acertos ",
                       "História", "% Acertos", "Física", "% Acertos", "Química", "% Acertos"]

#Arredondar as porcentagens para duas casas decimais
df_materias['% Acertos'] = np.round(df_materias['% Acertos'], 2)

# Exiba o DataFrame final
print(df_materias)
````

````python
# 2) (operações entre *DataFrames) Calcule o total de acertos de cada aluno.

df_total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui #somando os acertos por disciplina
df_total_acertos.rename(columns={'Qt_acertos': 'Total acertos'}, inplace=True) #Renomear a coluna para total acertos

df_total_acertos = df_total_acertos.drop(columns=['percentual_acerto'])

df_materias = (pd.concat([df_materias, df_total_acertos], axis = 1))# concatenando os dataframes


print(df_materias)
````

````python
# 3) Calcule o porcentual geral de cada aluno.

total_questoes = 24 + 18 + 8 + 8 + 8 + 12 + 12
porcentual_geral= round((df_total_acertos / total_questoes) * 100, 2) #Calcular o porcentual geral de acertos dos alunos

porcentual_geral.rename(columns={'Total acertos':'Porcentual geral'}, inplace=True)

df_materias = (pd.concat([df_materias, porcentual_geral], axis=1))



print(df_materias)
````

````python
# 4) Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?
alunos_acima_de_45 = (df_total_acertos >= 45).sum()
print(f'Quantidade de alunos que acertaram mais que 45 questões: \n{alunos_acima_de_45}')

df_materias['Segunda fase'] =  np.where( 
    (df_materias['Total acertos'] >= 45), 
    "Qualificado",
    "Desqualificado"
)

df_maior_45 = df_materias.loc[df_materias['Segunda fase'] == 'Qualificado']

df_maior_45.head(n=5)
````
