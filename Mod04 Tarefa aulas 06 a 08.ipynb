````python
# 1.1)
def consultar_dados_uf_tipo(uf, tipo_interessado):
   
    url = f"https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data['items'])
        return df
    else:
        print(f"Erro ao consultar os dados. Código de status: {response.status_code}")
        return None
````

````python
# 1.2)

# Exemplo de uso:
#uf_consulta = 'RJ'
#tipo_consulta = 'Município'
resultado = consultar_dados_uf_tipo('SP', 'Município')

if resultado is not None:
    print(resultado)
else:
    print("Erro ao obter os dados.")
````

````python
# 2) Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?
estado_MG = consultar_dados_uf_tipo('MG', 'Estado')
estado_MG_filtered = estado_MG[estado_MG['status'] == 'Arquivado por decurso de prazo']

status_MG = estado_MG_filtered['status'].value_counts()
status_MG
````

````python
# 3) Qual é o município da Bahia com mais solicitações deferidas?
municipios_BA = consultar_dados_uf_tipo('BA', 'Município')
municipios_BA.head()
````

````python
# 3.1)

solitacoes_municipio = municipios_BA['interessado'].value_counts()

print(solicitacoes_municipio)
print("")
print(f'O município da Bahia com mais solicitações deferidas foi: {solitacoes_municipio.index[0]} com {solitacoes_municipio.iloc[0]} solicitações.')
````

````python
# 4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
estado_BA = consultar_dados_uf_tipo('BA', 'Estado')
estado_BA.head(n=2)


estado_BA.to_csv('solicitacoes_bahia_estado.csv', index=False)

print("Arquivo CSV salvo com sucesso!")
````
