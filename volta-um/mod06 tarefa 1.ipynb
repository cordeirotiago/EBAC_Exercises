```python
#importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Carregar dataset sinasc_RO_2019
sinasc = pd.read_csv('b1abb33c4f.csv')

#Verificar dados
sinasc.head()
```

```python
#Verificar nome das colunas e o tipo dos dados
sinasc.info()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Quantidade de observações categorizadas pela classificação da variável APGAR1
sns.countplot(x = 'APGAR1', data = sinasc)
```

```python
#Quantidade de observações categorizadas pela classificação da variável APGAR5
sns.countplot(x = 'APGAR5', data = sinasc)
```

```python
# Quantidade de observações no dataset separados pelo SEXO
sns.countplot(x = 'SEXO', data = sinasc)
```

```python
#Quantidade de observações categorizadas pela quantidade de semanas de gestação
sns.countplot(x ='GESTACAO', data = sinasc)
plt.xticks(rotation=45)
```

```python
#Quantidade de observações categorizadas de acordo com o estado civil da mãe
sns.countplot(x = 'ESTCIVMAE', data = sinasc)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
#Nível de asfixia de nascidos vivos de acordo com a média do APGAR5 (quanto maior, menos asfixia) por consultas realizadas de pré-natal
sns.barplot(x='APGAR5', y= 'CONSULTAS', data = sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Observações categorizadas por diferentes variáveis de interesse para exploração dos dados em um mesmo painel

plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x = 'SEXO', data = sinasc)
sns.countplot(ax = axes[0, 1], x = 'GESTACAO', data = sinasc)
sns.countplot(ax = axes[1, 0], x = 'ESTCIVMAE', data = sinasc)
sns.countplot(ax = axes[1, 1], x = 'CONSULTAS', data = sinasc)

axes[0, 1].tick_params(axis = 'x', rotation=90)

plt.tight_layout()
plt.show()
```
